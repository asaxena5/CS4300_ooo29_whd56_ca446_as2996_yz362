```python
from bs4 import BeautifulSoup
import requests
import html5lib
import re
import pandas as pd
```

```python
URL = "https://healabel.com/carbon-footprint-of-foods"
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content, 'html5lib')
```

```python
soup
```

```python
d = {'Food':[],
    'CO2':[],
    'Footprint Level':[]}
```

```python
df = pd.DataFrame(d)
```

```python
sec1 = soup.find('div',{'id':'block-yui_3_17_2_1_1556431745228_512027'})
```

```python
sec2 = soup.find('div',{'id':'block-yui_3_17_2_1_1556431745228_620052'})
```

```python
sec3 = soup.find('div',{'id':'block-yui_3_17_2_1_1556431745228_634445'})
```

```python
sec4 = soup.find('div',{'id':'block-yui_3_17_2_1_1556431745228_603856'})
```

```python
for sec in [sec1, sec2, sec3, sec4]:
    for li in sec.findAll('li'):
        content = li.get_text() 

        if li.find('a') is not None:
            name = li.find('a').get_text()
        else:
            name = re.findall('^(.+?),', content)[0]


        co2 = float(re.findall("\d+\.\d+",content)[0])
        level = re.findall('([a-z]+) carbon footprint',content)[0]
        df.loc[len(df.index)] = [name,co2,level]
```

```python
df
```

```python
for food in df['Food'].unique():
    print([food, list(df['Food']).count(food)])
```

```python
df = df.drop_duplicates(subset='Food', keep='first')
```

```python
df
```

```python
df.to_csv('Food.com Recipes and Interactions/Footprint.csv',index=False)
```
